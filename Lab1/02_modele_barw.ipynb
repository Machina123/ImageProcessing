```python
from skimage import data

im = data.astronaut()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.imshow(im, cmap="gray")
plt.axis('off')
plt.show()
```

```python
print(im.shape, im.dtype)
```

```python
print(im[:5,:5])
```

```python
from skimage.color import convert_colorspace, rgb2lab, rgb2hsv, rgb2gray

imHSV = rgb2gray(im)
```

```python
print(imHSV.shape, imHSV.dtype, len(imHSV.shape))
```

```python
print(imHSV[:3,:3])
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.imshow(imHSV, cmap="gray")
plt.axis('off')
plt.show()
```
